```python
# Use a Chi-Squared statistical test for non-negative features to select 4 of the best features from the dataset
# Tutorial: https://www.datacamp.com/community/tutorials/feature-selection-python

import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Load the data into a Pandas Dataframe
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pd.read_csv(url, names=names)
print(dataframe)
print("")

# Get numpy arrays of the features (X) and target (Y)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]

# Feature extraction
test = SelectKBest(score_func=chi2, k=4)
print(type(X), type(Y))
fit = test.fit(X, Y)

# Summarize scores
np.set_printoptions(precision=3)
print(fit.scores_)

features = fit.transform(X)
# Summarize selected features
print(features[0:5,:])

# So the 4 attributes chosen (those with the highest scores): plas, test, mass, and age
```
